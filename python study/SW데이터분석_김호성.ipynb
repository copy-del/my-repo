```python
# 문제 4
# 다음 score_list에 들어있는 데이터를 이용해 아래와 같이 성적표를 만드는 프로그램을 작성하시오.
# (단, 학생의 숫자가 늘어나도 동작할 수 있도록 고려하시오.)
```

```python
score_list = [['김미희',89,78,99],
              ['서대희',67,88,77],
              ['강예진',100,55,96]]

col = ['이름','Python','Java','DB']

score_df=pd.DataFrame(score_list,columns =col)

score_mean = score_df[['Python','Java','DB']].mean(axis=1).round(1)

score_df['평균']=score_mean

score_df['학점']=''

def grade(x):
    if x>=90:
        result = 'A'
    elif x>=80:
        result = 'B'
    elif x>=70:
        result = 'C'
    elif x>=60:
        result = 'D'
    else:
        result = 'F'
    return result

for i in range(score_df['평균'].count()):
    score_df['학점'][i] = grade(score_df['평균'].values[i])

class_avg=score_df[['Python','Java','DB']].mean(axis=0).round(1)

score_df.loc['평균']=class_avg

score_df= score_df.fillna('')

score_df
```

```python
# 문제 6
#Traffic_Accident_2017.csv 데이터를 이용하여 
#‘차대 사람’인 사건 중 야간에 발생한 사건에 대해 아래 주어진 조건을 지켜 다음과 같이 시각화 하시오.
```

```python
data=pd.read_csv('LIB/data/Traffic_Accident_2017.csv',encoding = 'euc-kr')
data
```

```python
car_car=data[data['사고유형_대분류']=='차대사람']
```

```python
night=car_car[car_car['주야']=='야간']
```

```python
date=night['요일'].value_counts()
```

```python
date=date[['월','화','수','목','금','토','일']]
date
```

```python
x=date.index
y=date.values
```

```python
plt.bar(x,y)
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.ylim(0,200) # 축의 범위
plt.xlabel('요일') # x축 라벨
plt.ylabel('사고 건수') # y축 라벨

plt.show()
```
