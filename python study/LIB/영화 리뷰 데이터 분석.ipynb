#### 1. 데이터 불러오기

```python
# genfromtxt
# np.genfromtxt('파일경로', delimiter = '구분자::', dtype = 데이터 타입)

movie = np.genfromtxt('data/ratings.dat', delimiter ='::', dtype =np.int64)

movie
```

- 0번 인덱스 열 : 사용자의 id
- 1번 인덱스 열 : 영화의 id
- 2번 인덱스 열 : 영화의 평점
- 3번 인덱스 열 : 리뷰 입력 시간

#### 4. ID가 1인 사용자가 매긴 영화들의 평점들의 평균

```python
movie
```

```python
# boolean 인덱싱 : 특정 조건에 맞는 데이터를 가져올 때
# 아이디가 1인 사용자가 매긴 영화데이터
ID_1 = movie[:,0]==1
ID_1
```

```python
# 불리언 인덱싱
# ID가 1인 데이터를 가져오기

ID_1 = movie[movie[:,0]==1]
ID_1
```

```python
# ID가 1인 사람의 영화데이터'의' 평점을 가져오기

ID_1_rating = ID_1[:,2]
ID_1_rating
```

```python
# ID가 1인 사람의 영화데이터'의' 평점들'의' 평균
ID_1_rating.mean()
```

```python
# 중간에 변수로 만드는 과정 없이
movie[movie[:,0]==1][:,2].mean()
```

### 각 사용자가 매긴 평점들의 평균

```python
movie[movie[:,0]==1][:,2].mean()
movie[movie[:,0]==2][:,2].mean()
movie[movie[:,0]==3][:,2].mean()
movie[movie[:,0]==4][:,2].mean()
```

```python
# 데이터의 유효성 검사(6040가지 인지)
# 1~6040

# 중복값을 제거해주는(유일한 값들만 구해주는) 함수
# np.unique(시퀀스 자료형)

user_id = np.unique(movie[:,0])

user_id
```

```python
len(user_id)
```

```python
# 오름차순 sort()
np.sort(user_id)
```

```python
# 내림차순
np.sort(user_id)[::-1]
```

```python
# 전체 아이디의 수
user_id.size
```

```python
user_mean_list = []
for i in user_id:
    user_mean = movie[movie[:,0]==i][:,2].mean() # 사용자별로 영화 평점의 평균을 나타내는 user_mean
    user_mean_list.append([i,user_mean])
user_mean_list
```

### 6. 각 사용자별 평균 평점이 4점 이상인 사용자만 구해보기

```python
# 리스트 > 배열
# 조건문 적어서 불리언 인덱싱
```

```python
user_mean_arr = np.array(user_mean_list)

user_mean_arr
```

```python
# set_printoptions : numpy 배열 출력시 출력 형태를 설정하는 함수
# suppress=True : 소수점 자리수 고정 출력 선언
# precision=2 : 소수점 둘째짜리까지
np.set_printoptions(suppress=True, precision=2)
```

```python
user_mean_arr
```

```python
user_mean_arr[user_mean_arr[:,1]>=4]
```

```python
user_mean_4 = user_mean_arr[user_mean_arr[:,1]>=4]
user_mean_4
```

```python
user_mean_4[:,0]
```

```python
user_mean_arr[user_mean_arr[:,1]>=4][:,0].astype(int)
```
