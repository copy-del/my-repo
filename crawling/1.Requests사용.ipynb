```python
# res.text는 String 데이터
# ps를 이용해서 Sriing > html로 변환함.
from bs4 import BeautifulSoup as bs
```

```python
allNews = req.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query=%EC%9D%B4%EC%83%81%EA%B8%B0%ED%9B%84")
```

```python
allNews
```

```python
# (변환할 데이터, 파싱방법)
# 파싱(parsing) : 데이터를 특정 규칙에 맞게 구조화된 형태로 바꾸는 작업
soup = bs(allNews.text,"lxml")
```

```python
# 파싱 결과 : 객체 {태그:<span>, 속성:{align:center},콘텐츠:{"~~"}
# 컴퓨터가 각각의 키값을 통해서 값에 접근할 수 있다.
# 파싱을 하는 이유 : 컴퓨터에게 각각 태그, 속성, 컨텐츠에 접근 시키기 위해서.
```

```python
# select("선택자")
# HTME데이터에서 특정 요소(태그+콘텐츠)만 수집
# 명확하게 요청하지 않으면 정확한 수집이 불가능
news_title = soup.select("a.news_tit")
```

```python
news_title
```

```python
len(news_title)
```

```python
# 우리가 수집한 news_title 안에는 요소가 들어있음!
# 우리가 필요한건? content
# 해결법 수집된 요소에서 content만 뽑아내기
news_title[0].text
```

```python
# 반복문을 활용해서 모든 제목 데이터만 출력
# 활용을 위해서 제목 텍스트만 담을 리스트를 생성
title_list = []
for i in news_title :
    print(i.text)
    title_list.append(i.text)
```

```python
# pandas를 이용해 시각화
import pandas as pd
```

```python
data = pd.DataFrame(title_list, columns=["제목"])
data
```
