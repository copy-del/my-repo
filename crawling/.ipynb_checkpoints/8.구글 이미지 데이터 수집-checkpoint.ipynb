```python
# 스크롤을 이용해서 전체 이미지를 로딩
body = driver.find_element(By.CSS_SELECTOR,"body")
for i in range(10):
    body.send_keys(Keys.END)
    time.sleep(2)
```

```python
# 이미지 태그에서 src 속성만 추출
# 태그(요소)에서 속성을 꺼내올 때는 .get_attribute("속성명")
img = driver.find_elements(By.CSS_SELECTOR,".cC9Rib")
img[0].get_attribute("src")
len(img)
```

```python
img_src=[]
for i in range(len(img)):
    img_src.append(img[1].get_attribute("src"))
```

```python
img_src
```

```python
# src 값을 주면 실제 파일로 저장
from urllib.request import urlretrieve
```

```python
for i in range(len(img_src)):
    urlretrieve(img_src[i],"C:/Users/a/OneDrive/바탕 화면/이미지/"+str(i)+".jpg")
```
