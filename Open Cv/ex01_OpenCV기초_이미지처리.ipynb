### <색상공간 변환>

- 시각은 색상정보에 대해 잘못된 정보를 보내주는 경우가 많음
    - 밤에 색상을 볼 때 정확한 색상을 보여주지 못함 - 명도(검정색이 섞이는 것)
    - 색상에 빛을 강하게 비추는 경우, 정확한 색상을 보여주지 못함. -채도(흰색이 섞이는 것)
    - 컬러는 너무 많은 정보를 가지고 있기 때문에 연산량이 많음 - 실시간 처리가 느리다.
    - 따라서 상황에 따라, gray(흑,백,회색) 이미지나, binary(흑,백) 이미지로 변환해서 처리
    - gray 이미지 : 0~255로 된 픽셀로만 구성된 흑백이미지
    - binary(이진) 이미지 : 검정색(0)과 흰색(255)으로만 구성된 이미지, 0과 1로 이루어진 흑백이미지.
        - 특정 패턴이나 물체의 윤곽을 감지
        - 배경/물체 구별
        - 텍스트 필터링
        - 엑스레이/mri 스캔 사진
        - 자율주행시 장애물/차선 감지
        - 제품 결함 및 불량품 식별

```python
import matplotlib.pyplot as plt

img = cv2.imread('./data/pets.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img, cmap='gray')
plt.show()
```

```python
# 흑백 이미지로 읽어오기
img = cv2.imread('./data/ryan.jpg', cv2.IMREAD_GRAYSCALE) # 불러온 시점에서 RGB값 없음

# 이미지 이진화
# 임계값을 담을 변수, 변형된 이미지가 담길 변수 = cv2.threshold(대상, 임계값(기준값), 값, 이진화종류값)
_, thr_img = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)

# 오츠 이진화
th_val, thr_img2 = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

# plt.subplot(행, 열, 순서)
plt.subplot(1, 3, 1) # 1행 3열 중, 첫번째 자리 // 그리드와 유사
plt.title(f'orginal')
plt.axis('off')
plt.imshow(img, cmap='gray')

# plt.subplot(행, 열, 순서)
plt.subplot(1, 3, 2) # 1행 3열 중, 두번째 자리 // 그리드와 유사
plt.title(f'binary:{_}')
plt.axis('off')
plt.imshow(thr_img, cmap='gray')

plt.subplot(1, 3, 3) # 1행 3열 중, 세번째 자리 // 그리드와 유사
plt.title(f'otsu binary:{th_val}')
plt.axis('off')
plt.imshow(thr_img2, cmap='gray')
```

```python
# 컬러로 가져온 후 흑백 사진으로 변환
apple = cv2.imread('./data/apple.jpg', cv2.IMREAD_COLOR)
apple_gray = cv2.cvtColor(apple, cv2.COLOR_BGR2GRAY)

height, width = apple_gray.shape
center_x, center_y = width//2, height//2 # 306, 306

# 정사각형 영역 접근
# 좌상단
top_left_x, top_left_y = center_x//2, center_y//2 # 153, 153

# 우하단
bottom_right_x, bottom_right_y = center_x + top_left_x, center_y + top_left_y

# 사각형 영역 추출하기 - 배열 슬라이싱
# 배열[행의 범위, 열의 범위]
rect_region = apple_gray[top_left_y:bottom_right_y, top_left_x:bottom_right_x] # [ 153:459 , 153:459 ]
print(rect_region.shape)

plt.imshow(rect_region, cmap='gray')
plt.show()
```

```python
# 이진화
val, apple_bin = cv2.threshold(rect_region, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
plt.imshow(apple_bin, cmap='gray')
plt.show()
```

```python
# 원본 이미지에 이진화 변형 이미지 대입
# 복사 먼저 진행
apple2 = apple_gray.copy() # 깊은 복사(원본 형태 복사) : 원본이 변형되더라도 그대로 유지됨.

apple2[top_left_y:bottom_right_y, top_left_x:bottom_right_x] = apple_bin # 이진화된 값으로 변경
plt.imshow(apple2, cmap='gray')
plt.show()
```

### 이미지 블랜딩

```python
# 이미지 2개 선택해서 불러오기
# 2장의 이미지 출력하기

img1 = cv2.imread('./data/ryan.jpg', cv2.IMREAD_COLOR)
img2 = cv2.imread('./data/apple.jpg', cv2.IMREAD_COLOR)

# 색상공간 변환 bgr > rgb
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# 이미지 크기 변경 - 사이즈가 동일해야 블렌딩이 가능.
img1 = cv2.resize(img1, dsize=(300,300))
img2 = cv2.resize(img2, dsize=(300,300))
print(img1.shape)
print(img2.shape)

# 두 이미지 블랜딩 cv2.addWeighted(이미지1, 1가중치, 이미지2, 2가중치, 전체가중치
# 두개의 이미지로 새로운 이미지(배열)을 만드는 공식에 대입되는 가중치 설정
result = cv2.addWeighted(img1, 0.5, img2, 0.5, 1)

plt.subplot(1,3,1) # 위치 잡고
plt.imshow(img1) # 그립 넣고
plt.subplot(1,3,2)
plt.imshow(img2)
plt.subplot(1,3,3)
plt.imshow(result)
```
