```python
import cv2
import matplotlib.pyplot as plt
```

```python
# cv2 이미지를 출력하는 기능 실행
img = cv2.imread('./data/pets.jpg')
cv2.imshow('pet', img)

# 창 닫힘 기능
print(cv2.waitKey(0)) # 0 : 무한대기, ms초 단위로 값을 입력 1s = 1000
cv2.destroyAllWindows()
```

### 동영상 출력하기
- cv2.waitKey(밀리초): 동영상이 실행되는 시간에 대해서 제어
- cv2.waitKey(): 사용자의 키보드로부터 입력된 아스키 코드값을 반환 > 창이 꺼지는 기능

```python
cap = cv2.VideoCapture(0) # 영상(비디오 객체) 가져옴
cnt = 0
while True: # 계속 실행, 동영상 : 연속적인 이미지
    ch, frame = cap.read() # ch : 성공여부(True or False), frame : 받아온 이미지

    # 실패할 경우 동영상 멈추기 while break
    if not ch:
        print('동영상 재생 끝 or 실패')
        break
    # 제대로 읽었을 경우
    cv2.imshow('video', frame)

    # 동영상 종료 키 지정
    key = cv2.waitKey(15) # 속도 제어. >> 0 은 무한대기라서 재생 X
    if key == ord('q'): # ord() : 아스키코드 값을 출력하는 함수
        print('사용자에 의해서 영상이 종료')
        break

    # 사진 찍기, 프레임 저장하기
    if key == ord('w'):
        cnt += 1
        print(f'./data/hoseong{cnt}.png 사진저장')
        cv2.imwrite(f'./data/hoseong{cnt}.png', frame, params=[cv2.IMWRITE_PNG_COMPRESSION, 0])
        # imwrite : 이미지 저장 함수( 저장경로, 저장 이미지, 저장시 옵션(압축정도, 0일경우 최고화질))
        

cap.release() # 카메라 연결 자원 해제
cv2.destroyAllWindows()
```
