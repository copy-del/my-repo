```python
# 피클 임시저장용

import pickle

def save_pickle(obj, path: str):
    with open(path, "wb") as f:
        pickle.dump(obj, f)
    print(f"✅ 저장 완료 → {path}")

def load_pickle(path: str):
    with open(path, "rb") as f:
        obj = pickle.load(f)
    print(f"✅ 로드 완료 ← {path}")
    return obj
```

```python
import itertools
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

def embed_and_compare(
    account_list: list[str],
    model_name: str = "snunlp/KR-SBERT-V40K-klueNLI-augSTS",
    top_k: int = 50
) -> pd.DataFrame:
    """
    account_list: 계정명 리스트
    model_name: 불러올 SentenceTransformer 모델
    top_k: 유사도 상위 N개 페어 반환
    """
    print(f"🔍 모델 로딩: {model_name}")
    model = SentenceTransformer(model_name)

    print(f"🔢 임베딩 중... ({len(account_list)}개)")
    embs = model.encode(account_list, convert_to_numpy=True)

    print(f"📐 유사도 계산...")
    sim = cosine_similarity(embs)
    pairs = [
        (account_list[i], account_list[j], sim[i, j])
        for i, j in itertools.combinations(range(len(account_list)), 2)
    ]
    top_pairs = sorted(pairs, key=lambda x: x[2], reverse=True)[:top_k]
    df = pd.DataFrame(top_pairs, columns=["Account1", "Account2", "Similarity"])
    return df
```

```python
def match_with_targets(
    all_accounts: list[str],
    target_keywords: list[str],
    model_name: str = "snunlp/KR-SBERT-V40K-klueNLI-augSTS",
    top_k: int = 5
) -> pd.DataFrame:
    print("📦 모델 로딩 중...")
    model = SentenceTransformer(model_name)

    print("🔢 전체 계정명 임베딩 중...")
    all_embs = model.encode(all_accounts, convert_to_numpy=True)

    print("🎯 타겟 키워드 임베딩 중...")
    target_embs = model.encode(target_keywords, convert_to_numpy=True)

    results = []

    for i, keyword in enumerate(target_keywords):
        sims = cosine_similarity([target_embs[i]], all_embs)[0]
        top_indices = sims.argsort()[::-1][:top_k]

        for idx in top_indices:
            results.append((keyword, all_accounts[idx], sims[idx]))

    df_result = pd.DataFrame(results, columns=["TargetKeyword", "MatchedAccount", "Similarity"])
    return df_result.sort_values(by=["TargetKeyword", "Similarity"], ascending=[True, False])
```

```python
def normalize_account_name(name: str) -> str:
    name = name.replace(" ", "")
    name = name.replace("-", "")
    name = name.replace("(", "").replace(")", "")
    name = name.lower()
    return name
```

```python
# 전체 df 통합
df_all = pd.concat(account_rows, ignore_index=True)
# 중복제거
unique_accounts = df_all['account_nm'].dropna().unique().tolist()
# 전처리로 단순 표기 통합
df_all['normalized'] = df_all['account_nm'].fillna("").apply(normalize_account_name)
print(f"\n총 고유 계정명 수: {len(unique_accounts)}")
print(len(df_all['normalized']))
```

```python


# ✅ Step 0: 컬럼 정리
corp_columns = unique_accounts+ [
    ['매출액', '영업이익', '순이익'],
    ['매출', '영업수익', '당기순이익'],
    ['Revenue', 'OperatingProfit', 'NetIncome'],
    ['총매출액', '영업이득', 'Net Income'],
    ['매출총액', '영업이익', '순이익']
]

import re

def clean_name(name):
    # 한글, 알파벳, 숫자만 남기고 나머지 제거
    name = re.sub(r"[^가-힣a-zA-Z0-9]", "", name.lower())
    return name


all_columns = [clean_name(col) for corp in corp_columns for col in corp]
unique_columns = list(set(all_columns))


# ✅ Step 1: 컬럼명 임베딩 (의미 분석)
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
embeddings = model.encode(unique_columns)


# ✅ Step 2: 클러스터링 (비슷한 컬럼끼리 그룹)
clustering = DBSCAN(eps=0.5, min_samples=2, metric='cosine').fit(embeddings)

cluster_map = {}
for idx, label in enumerate(clustering.labels_):
    if label == -1:
        continue
    cluster_map.setdefault(label, []).append(unique_columns[idx])

# ✅ Step 3: 전체 컬럼 빈도 수 세기
column_counter = Counter(all_columns)


# ✅ Step 4: 각 그룹에서 가장 많이 쓰인 컬럼 1개 대표로 뽑기
cluster_representatives = {}
for label, cols in cluster_map.items():
    most_common = sorted(cols, key=lambda x: column_counter[x], reverse=True)[0]
    cluster_representatives[most_common] = cols


# 매핑
account_mapping = {}
for rep, synonyms in cluster_representatives.items():
    for s in synonyms:
        account_mapping[s] = rep


# ✅ Step 5: 대표 컬럼 중에서 사용 빈도 높은 순으로 정렬
rep_counter = {
    rep: sum(column_counter[c] for c in synonyms)
    for rep, synonyms in cluster_representatives.items()
}


top_features = sorted(rep_counter.items(), key=lambda x: x[1], reverse=True)[:50]


# ✅ 결과 보기
import json
for feature, freq in top_features:
    print(f"[{feature}] (빈도: {freq}) → 유사 컬럼들: {json.dumps(cluster_representatives[feature], ensure_ascii=False)}")
```

```python
df_all['normalized'] = df_all['account_nm'].fillna('').apply(clean_name)
df_all['mapped'] = df_all['normalized'].map(account_mapping)
```
